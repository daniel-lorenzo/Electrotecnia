## Problema 6
Un elemento de circuito tiene:

$\begin{array}{l}
i(t) = 13,45 \sin (500 t - 53,4^\circ) \, \mathrm{A} \\
v(t) = 150 \sin (500 t + 10^\circ) \, \mathrm{V}
\end{array}$

Determinar la característica del elemento.

<img src="img/tp2e5.png" alt="tp2e5.png">

### Solución

Se representan los fasores equivalentes de las funciones del tiempo en el dominio de la frecuencia.

Los fasores equivalentes de las funciones del tiempo de la tensión y la corriente resultan,

$$ \vec{I} = I_\mathrm{ef} \angle 53,4^\circ = \frac{ I_\mathrm{m \acute ax}}{\sqrt{2}} e^{-j53,4^\circ}  $$

$$ \vec{V} = V_\mathrm{ef} \angle 10^\circ = \frac{V_\mathrm{m \acute ax}}{\sqrt{2}} e^{j10^\circ} $$

La diferencia de fase o desfase entre las ondas de tensión y corriente viene dada por:

$$ \theta = \varphi_v - \varphi_i = 10^\circ - (-53,40^\circ) = 63,4^\circ $$

En base al ángulo de desfase podemos decir que se trata de un elemento resistivo-inductivo, ya que la tensión adelanta a la corriente en 63,4° y por lo tanto se verifica que la impedancia del circuito eléctrico viene dada por:

$$ Z = \frac{\vec{V}}{\vec{I}} = \frac{V_\mathrm{ef} e^{j10^\circ}}{I_\mathrm{ef} e^{-j53^\circ}} = \frac{V_\mathrm{m \acute ax}}{I_\mathrm{m \acute ax}} e^{j(10^\circ + 53,4^\circ)} $$

$$ Z = \frac{150}{13,45} e^{j63,4^\circ} \, \Omega = 11,15 e^{j63,4^\circ} \, \Omega $$

Además, se sabe que:

$$ Z = |Z| e^{j \varphi} $$

Entonces:
$$ |Z| = 11,15 \, \Omega \qquad \wedge \qquad \varphi = 63,4^\circ $$

y que, además:

$$ Z = R + jX_L = |Z| ( \cos \varphi + j \sin \varphi ) $$

$$ Z = 11,15 \, \Omega (\cos 63,4^\circ + j \sin 63,4^\circ) = (4,99 + j9,97) \, \Omega $$

El resultado obtenido en la ecuación anterior implica que $R = 4,99 \, \Omega$ y que $X_L = 9,97 \, \Omega$.

Pero, como además $X_L = \omega L$, resulta que la inductancia del circuito es:

$$ L = \frac{X_L}{\omega} = \frac{9,97 \, \Omega}{500 \, \mathrm{s}^{-1}} = 0,01994 \, \mathrm{Hy} $$

```python
import math, cmath
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.figure(figsize=(6,6))
phi_i = math.radians(-53.4)
phi_v = math.radians(10)
r_i = 13.45/math.sqrt(2)
r_v = 150/math.sqrt(2)

plt.polar(phi_i , r_i, 'bo')
plt.text(phi_i , r_i + 5, r'$I$', fontsize=18)
plt.polar(phi_v, r_v, 'ro')
plt.text(phi_v , r_v + 5, r'$V$', fontsize=18)

plt.polar([0,phi_i],[0,r_i],linewidth=3)
plt.polar([0,phi_v],[0,r_v],linewidth=3)
plt.grid(linestyle="--")
plt.show()
```

```python
I = cmath.rect( 13.45 / math.sqrt(2) , math.radians(-53.4) ) 
V = cmath.rect( 150 / math.sqrt(2) , math.radians(10) )

Z = V/I
Zpol = cmath.polar(Z)
```

```python
%matplotlib inline
plt.figure(figsize=(6,6))

plt.polar( Zpol[1] , Zpol[0] , 'bo')
plt.polar([0, Zpol[1] ],[0, Zpol[0] ],linewidth=3)
plt.text(Zpol[1] , Zpol[0] + 0.5, r'$Z$', fontsize=18)
plt.grid(linestyle="--")
plt.show()
```

```python
print('Z = {:.3f}'.format(Z))
print('Zpol = (%.2f < %.2f°)'%(Zpol[0] , math.degrees(Zpol[1]) ))
```
