## Ejermplo 12.1
Determine la secuencia de fases del conjunto de tensiones

$$\begin{array}{l}
v_{an} = 200 \cos (\omega t + 10^\circ ) \\
v_{bn} = 200 \cos (\omega t - 230^\circ) \\
v_{cn} = 200 \cos (\omega t - 110^\circ)
\end{array}$$

### Solución
Las tensiones pueden expresarse en forma fasorial como

$$ V_{an} = 200 \angle 10^\circ \, \mathrm{V}, \qquad V_{bn} = 200 \angle -230^\circ \, \mathrm{V}, \qquad V_{cn} = 200 \angle -110^\circ \, \mathrm{V} $$

```python
import math, cmath
import numpy as np
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='polar')

ax.set_xticks(np.arange(0,2.0*np.pi,np.pi/6.0))

phi_Van = math.radians(10)
phi_Vbn = math.radians(-230)
phi_Vcn = math.radians(-110)

r_V = 200

plt.polar(phi_Van, r_V, 'ko')
plt.text(phi_Van , r_V + 70 , r'$V_{an}$', fontsize=18)
plt.polar(phi_Vbn, r_V, 'ko')
plt.text(phi_Vbn , r_V + 70 , r'$V_{bn}$', fontsize=18)
plt.polar(phi_Vcn, r_V, 'ko')
plt.text(phi_Vcn , r_V + 70 , r'$V_{cn}$', fontsize=18)

plt.polar([0,phi_Van],[0,r_V],linewidth=3)
plt.polar([0,phi_Vbn],[0,r_V],linewidth=3)
plt.polar([0,phi_Vcn],[0,r_V],linewidth=3)

plt.show()
```

Es notorio que $V_{an}$ se adelanta a $V_{cn}$ en 120°, y que $V_{cn}$ se adelanta a su vez a $V_{bn}$ en 120°. Así, se tiene una secuencia $acb$.

## Conexión estrella-estrella balanceada

<div class="alert-info">Un sistema Y-Y balanceado es un sistema trifásico con fuente balanceada conectada en Y y carga balanceada conectada en Y.
    </div>

## Ejemplo 12.2
Calcule las corrientes de línea del sistema Y-Y de tres conductores de la figura.

<img src="img/ej12-2.png">

### Solución
El circuito trifásico de la figura está balanceado. $I_a$ se obtiene del análisis monofásico como

$$ I_a = \frac{V_{an}}{Z_Y} $$

```python
import math, cmath
```

```python
Zy = 5-2j + 10+8j
print('Zy = {:.0f} Ohm'.format(Zy))
```

```python
Van = 110
Ia = Van/Zy
```

Como las tensiones de fuente de la figura están en secuencia positiva, las corrientes de línea también están en secuencia positiva:

$$ I_b = I_a \angle -120^\circ $$

$$ I_c = I_a \angle -240^\circ $$

```python
Ib = cmath.rect(abs(Ia) , cmath.phase(Ia) - math.radians(120) )
Ic = cmath.rect(abs(Ia) , cmath.phase(Ia) - math.radians(240) )
```

```python
print('Corrientes de línea:')
print('Ia = (%.2f < %.2f°) A'%(abs(Ia) , math.degrees( cmath.phase(Ia) ) ))
print('Ib = (%.2f < %.2f°) A'%(abs(Ib) , math.degrees( cmath.phase(Ib) ) ))
print('Ic = (%.2f < %.2f°) A'%(abs(Ic) , math.degrees( cmath.phase(Ic) ) ))
```

## Problema de práctica 12.2
Un generador trifásico balanceado conectado en Y con una impedancia de $0,4 + j0,3 \, \Omega$ por fase se conecta con una carga balanceada conectada en Y con una impedancia de $24 + j19 \, \Omega$ por fase. La línea que une al generador y la carga tiene una impedancia de $0,6 + j0,7 \, \Omega$ por fase. Suponiendo una secuencia positiva de las tensiones de fuente y que $V_{an} = 120 \angle 30^\circ \, \mathrm{V}$ halle:
1. Las tensiones de línea.
2. Las corrientes de línea.

### Solución

<img src="img/ejp12-2.png">

$\begin{array}{l}
\varphi_a = 30^\circ \\
\varphi_b = 30 - 120^\circ = -90^\circ \\
\varphi_c = 30 + 120^\circ = 150^\circ
\end{array}$

__Tensiones de línea:__

Datos:
$\left\{
\begin{array}{l}
V_{an} = 120 \angle 30^\circ \, \mathrm{V} \\
V_{bn} = 120 \angle -90^\circ \, \mathrm{V} \\
V_{cn} = 120 \angle 150^\circ \, \mathrm{V}
\end{array}
\right.$

$$ V_p = |V_{an}| = |V_{bn}| = |V_{cn}| $$

$$ |V_L| = V_p \sqrt{3} $$

$$ |V_L| = 120 \sqrt{3} \cong 207,85 \, \mathrm{V} $$

__Tensiones de fase:__

$\begin{array}{l}
V_{ab} = V_{an} \sqrt{3} \angle ( \varphi_a + 30^\circ ) = 207,85 \angle 60^\circ \, \mathrm{V} \\
V_{bc} = V_{bn} \sqrt{3} \angle ( \varphi_b + 30^\circ ) = 207,85 \angle -60^\circ \, \mathrm{V} \\
V_{ca} = V_{cn} \sqrt{3} \angle ( \varphi_c + 30^\circ ) = 207,85 \angle 180^\circ  \, \mathrm{V} \\
\end{array}$

```python
import math, cmath
```

```python
Vp = 120
phi = 30
Van = cmath.rect(Vp, math.radians(phi) ) # V
Vbn = cmath.rect(Vp, math.radians(phi-120) ) # V
Vcn = cmath.rect(Vp, math.radians(phi+120) ) # V

V_lin = Vp*math.sqrt(3)

Vab = cmath.rect(V_lin, math.radians(phi + 30) )
Vbc = cmath.rect(V_lin, math.radians(phi-120+30) )
Vca = cmath.rect(V_lin, math.radians(phi+120+30) )
```

```python
print('Tensiones de línea:')
print('Vab = (%.2f < %.2f°) V'%(abs(Vab), math.degrees( cmath.phase(Vab) ) ) )
print('Vbc = (%.2f < %.2f°) V'%(abs(Vbc), math.degrees( cmath.phase(Vbc) ) ) )
print('Vca = (%.2f < %.2f°) V'%(abs(Vca), math.degrees( cmath.phase(Vca) ) ) )
```

Diagramas fasoriales que ilustran la relación entre tensiones de línea y las de fase.

<img src="img/ejp12-2-2.png">

<img src="img/ejp12-2-3.png">

$$ V_{an} = I_a Z_Y $$ 

Entonces
$$ I_{an} = \frac{V_{an}}{Z_Y} = \frac{120 \angle 30^\circ}{25 + j20} = 3,75 \angle -8,66^\circ \, \mathrm{A} $$

$$ I_{bn} = \frac{V_{bn}}{Z_Y} = \frac{120 \angle -90^\circ}{25 + j20} = 3,75 \angle -128,66^\circ \, \mathrm{A} $$

$$ I_{cn} = \frac{V_{cn}}{Z_Y} = \frac{120 \angle 150^\circ}{25 + j20} = 3,75 \angle 111,34^\circ \, \mathrm{A} $$

```python
Zy = (0.4+0.3j) + (0.6+0.7j) + (24+19j) # Ohm

Ian = Van/Zy
Ibn = Vbn/Zy
Icn = Vcn/Zy
```

```python
print('Corrientes de línea:')
print('Ian = (%.2f < %.2f°) A'%(abs(Ian) , math.degrees( cmath.phase(Ian) ) ))
print('Ibn = (%.2f < %.2f°) A'%(abs(Ibn) , math.degrees( cmath.phase(Ibn) ) ))
print('Icn = (%.2f < %.2f°) A'%(abs(Icn) , math.degrees( cmath.phase(Icn) ) ))
```

## Conexión estrella-delta balanceada

<div class="alert-info">
Un sistema $Y-\Delta$ balanceado consta de una fuente balanceada conectada en
Y que alimenta a una carga balanceada conectada en $\Delta$.
    </div>

## Ejemplo 12.3
Una fuente balanceada conectada en Y es secuencia *abc* con $V_{an} = 100 \angle 10^\circ \, \mathrm{V}$ se conecta con una carga balanceada en $\Delta$ de $(8+j4) \, \Omega$ por fase. Calcule las corrientes de fase y de línea.

### Solución

<img src="img/ej12-3.png">

### Método 1

$$ V_{ab} = V_{an} \sqrt{3} \cdot (1 \angle 30^\circ) = 173,21 \angle 40^\circ \, \mathrm{V} $$

Las corrientes de f son

$$ I_{ab} = \frac{V_{ab}}{Z_\Delta} = 19,36 \angle 13,43^\circ \, \mathrm{A} $$

$$ I_{bc} = I_{ab} \cdot (1 \angle -120^\circ) \, \mathrm{A} $$

$$ I_{ca} = I_{ab} \cdot (1 \angle +120^\circ) \, \mathrm{A} $$

Las corrientes de línea son

$$ I_a = I_{ab} \sqrt{3} \cdot (1 \angle -30^\circ) \, \mathrm{A} $$

$$ I_b = I_a \cdot (1 \angle -120^\circ) \, \mathrm{A} $$

$$ I_c = I_a \cdot (1 \angle + 120^\circ) \, \mathrm{A} $$

```python
import math, cmath
```

```python
Van = cmath.rect(100, math.radians(10) )
ZD = 8 + 4j
Vab = Van*math.sqrt(3) * cmath.rect(1, math.radians(30) )
# Las corrientes de fase son
Iab = Vab/ZD
Ibc = Iab * cmath.rect(1 , math.radians(-120) )
Ica = Iab * cmath.rect(1 , math.radians(120) )
```

```python
print('Vab = (%.2f < %.2f°) V'%(abs(Vab), math.degrees( cmath.phase(Vab) ) ))
```

```python
print('Corrientes de fase:')
print('Iab = (%.2f < %.2f°) A'%(abs(Iab) , math.degrees( cmath.phase(Iab) ) ))
print('Ibc = (%.2f < %.2f°) A'%(abs(Ibc) , math.degrees( cmath.phase(Ibc) ) ))
print('Ica = (%.2f < %.2f°) A'%(abs(Ica) , math.degrees( cmath.phase(Ica) ) ))
```

```python
# Corrientes de línea
Ia = Iab*math.sqrt(3) / cmath.rect(1, math.radians(30) )
Ib = Ia * cmath.rect(1 , math.radians(-120))
Ic = Ia * cmath.rect(1 , math.radians(120))
```

```python
print('Corrientes de línea:')
print('Ia = (%.2f < %.2f°) A'%(abs(Ia) , math.degrees( cmath.phase(Ia) ) ))
print('Ib = (%.2f < %.2f°) A'%(abs(Ib) , math.degrees( cmath.phase(Ib) ) ))
print('Ic = (%.2f < %.2f°) A'%(abs(Ic) , math.degrees( cmath.phase(Ic) ) ))
```

### Método 2

Alternativamente, aplicando el análisis monofásico

$$ I_a = \frac{V_{an}}{Z_\Delta / 3} $$

```python
I_a = Van/(ZD/3)

print('I_a = (%.2f < %.2f°) A'%(abs(I_a) , math.degrees( cmath.phase(I_a) ) ))
```

como se obtuvo anteriormente. Las demás corrientes de línea se obtienen siguiendo la secuencia de fases *abc*.

```python
print('I_b = (%.2f < %.2f°) A'%(abs(I_a) , math.degrees( cmath.phase(I_a) ) -120 ))
print('I_c = (%.2f < %.2f°) A'%(abs(I_a) , math.degrees( cmath.phase(I_a) ) +120 ))
```

## Conexión delta-delta balanceada

<div class="alert-info">
Un sistema $\Delta - \Delta$ balanceado es aquel en el que tanto la fuente balanceada
como la carga balanceada están conectadas en $\Delta$.
    </div>

## Ejemplo 14.4
Una carga balanceada en $\Delta$ y con impedancia $20 - j15 \, \Omega$ se conecta con un generador conectado en $\Delta$ eb secuencia positiva con $V_{ab} = 330^\circ \, \mathrm{V}$. Calcule las corrientes de fase de la carga y las corrientes de línea.

<img src="img/ej12-4.png">

### Solución
La impedancia de carga por fase es

$$ Z_\Delta = 20 - j15 \, \Omega $$

```python
import math, cmath
```

```python
# Datos:
ZD = 20-15j # Ohm
Vab = 330 # V
```

```python
Iab = Vab/ZD
# Resto 120° a Iab
Ibc = Iab * cmath.rect(1, math.radians(-120) )
# Sumo 120° a Iab
Ica = Iab * cmath.rect(1, math.radians( 120) )
```

```python
print('Corrientes de fase:')
print('Iab = (%.2f < %.2f°) A'%(abs(Iab) , math.degrees( cmath.phase(Iab) ) ))
print('Ibc = (%.2f < %.2f°) A'%(abs(Ibc) , math.degrees( cmath.phase(Ibc) ) ))
print('Ica = (%.2f < %.2f°) A'%(abs(Ica) , math.degrees( cmath.phase(Ica) ) ))
```

En el caso de una carga en delta, la corriente de línea se atrasa de la correspondiente corriente de fase en 30° y tiene una magnitud de $\sqrt{3}$ veces la de la corriente de fase. En consecuencia, las corrientes de línea son

$$ I_a = I_{ab} \sqrt{3} \cdot (1 \angle -30^\circ) \, \mathrm{A} $$

$$ I_b = I_a \cdot (1 \angle - 120^\circ) \, \mathrm{A} $$

$$ I_c = I_a \cdot (1 \angle 120^\circ ) \, \mathrm{A} $$

```python
Ia = Iab*math.sqrt(3) * cmath.rect(1 , math.radians(-30) ) # A
Ib = Ia * cmath.rect(1 , math.radians(-120)) # A
Ic = Ia * cmath.rect(1 , math.radians(120)) # A
```

```python
print('Corrientes de línea:')
print('Ia = (%.2f < %.2f°) A'%(abs(Ia) , math.degrees( cmath.phase(Ia) ) ))
print('Ib = (%.2f < %.2f°) A'%(abs(Ib) , math.degrees( cmath.phase(Ib) ) ))
print('Ic = (%.2f < %.2f°) A'%(abs(Ic) , math.degrees( cmath.phase(Ic) ) ))
```

## Ejemplo 12.5
Una carga balanceada en Y con una impedancia de fase de $40 + j25 \, \Omega$ se alimenta con una fuente balanceada conectada en $\Delta$ en secuencia positiva con una tensión de línea de 210 V. Calcule las corriente de fase. Use $V_{ab}$ como referencia.

<img src="img/ej12-5.png">

### Solución
__Datos:__
$\left\{
\begin{array}{l}
Z_Y = 40+25j \, \Omega \\
V_{ab} = 210 \, \mathrm{V}
\end{array}
\right.$

Cuando la fuente conectada en $\Delta$ se transforma en una fuente conectada en Y,

$$ V_{an}= \frac{V_{ab}}{\sqrt{3}} \cdot (1 \angle -30^\circ) $$

Las corrientes de línea son

$$ I_a = \frac{V_{an}}{Z_Y} $$

$$ I_b = I_a \cdot (1 \angle -120^\circ) $$

$$ I_c = I_a \cdot (1 \angle 120^\circ) $$

```python
import math, cmath
```

```python
# Datos:
Zy = 40+25j # Ohm
Vab = 210 # V
```

```python
Van = Vab/math.sqrt(3) * cmath.rect(1 , math.radians(-30) )
```

```python
print('Van = (%.2f < %.2f°) V'%(abs(Van) , math.degrees( cmath.phase(Van) ) ))
```

```python
Ia = Van/Zy
Ib = Ia * cmath.rect(1 , math.radians(-120) )
Ic = Ia * cmath.rect(1 , math.radians( 120) )
```

```python
print('Corrientes de línea:')
print('Ia = (%.2f < %.1f°) A'%(abs(Ia) , math.degrees( cmath.phase(Ia) ) ))
print('Ib = (%.2f < %.1f°) A'%(abs(Ib) , math.degrees( cmath.phase(Ib) ) ))
print('Ic = (%.2f < %.1f°) A'%(abs(Ic) , math.degrees( cmath.phase(Ic) ) ))
```

## Problema de práctica 12.5
En un circuito $\Delta - Y$ balanceado, $V_{ab} = 240 \angle 15^\circ$ y $Z_Y = (12 + j15) \, \Omega$. Calcule las corrientes de línea.

<img src="img/ej12-5.png">

### Solución
__Datos:__
$\left\{
\begin{array}{l}
V_{an} = 240 \angle 15^\circ \\
Z_Y = (12 + j15) \, \Omega
\end{array}
\right.
$

Cuando la fuente conectada en $\Delta$ se transforma en una fuente conectada en Y,

$$ V_{an} = \frac{V_{ab}}{\sqrt{3}} \cdot (1 \angle -30^\circ) $$

```python
import math, cmath
```

```python
# Datos:
Vab = cmath.rect(240 , math.radians(15) ) # V
Zy = 12+15j # Ohm
```

Las corrientes de línea son

```python
Ia = Van/Zy
Ib = Ia * cmath.rect(1 , math.radians(-120 ))
Ic = Ia * cmath.rect(1 , math.radians( 120 ))
```

```python
print('Van = (%.2f < %.2f°) V'%(abs(Van) , math.degrees( cmath.phase(Van) ) ))
print('Corrientes de línea:')
print('Ia = (%.2f < %.2f°) A'%(abs(Ia) , math.degrees( cmath.phase(Ia) ) ))
print('Ib = (%.2f < %.2f°) A'%(abs(Ib) , math.degrees( cmath.phase(Ib) ) ))
print('Ic = (%.2f < %.2f°) A'%(abs(Ic) , math.degrees( cmath.phase(Ic) ) ))
```
