### Solución

Malla 1
$$ -V_1 + R_1 I_1 + R_2 (I_1 - I_3) = 0 $$

$$ - 10 + 20 I_1 + 100 (I_1 - I_3) = 0  $$

$$ 120 I_1 - 100 I_3 = 10 \tag{1} $$

Malla 2
$$ V_2 + R_2 (I_3 - I_1) + R_3 I_3 = 0 $$

$$ 5 + 100 (I_3 - I_1) + 10 I_3 = 0 $$

$$ 100 I_1 - 110 I_3 = 5 \tag{2} $$

__En forma matricial__

$$\left[

\right]
\left[
\begin{array}{c}

### Método 1 - Aplicando regla de Cramer
Calculamos los determinantes

$$\Delta =
\left|
\begin{array}{cc}
120 & -100 \\
100 & -110
\end{array}
\right|$$

$$\Delta_1 =
\left|
\begin{array}{cc}
10 & -100 \\
5 & -110
\end{array}
\right|$$

$$\Delta_3 =
\left|
\begin{array}{cc}
120 & 10 \\
100 & 5
\end{array}
\right|$$

```python
import numpy as np
```

```python
M = np.array( [ [120 , -100] , [100 , -110] ] )
M1 = np.array( [ [10 , -100] , [5 , -110] ] )
M3 = np.array( [ [120 , 10] ,  [100 , 5] ] )
Δ = np.linalg.det(M)
Δ1 = np.linalg.det(M1)
Δ3 = np.linalg.det(M3)
```

```python
#@title resultados:
print('Δ = %.1f'%Δ)
print('Δ1 = %.1f'%Δ1)
print('Δ3 = %.1f'%Δ3)
```

$$ I_1 = \frac{\Delta_1}{\Delta} \quad ; \quad I_3 = \frac{\Delta_3}{\Delta} $$

```python
I1 = Δ1/Δ
I3 = Δ3/Δ
I2 = I1 - I3
```

```python
#@title resultados
print('Resumen:')
print('I1 = %.4f [A] = %.1f [mA]'%(I1 , I1*1000) )
print('I2 = %.4f [A] = %.1f [mA]'%(I2 , I2*1000) )
print('I3 = %.4f [A] = %.1f [mA]'%(I3 , I3*1000) )
```

### Método 2
$I_1$ e $I_3$ pueden calcularse también haciendo:

$$ A = 
\left[
\begin{array}{cc}
120 & -100 \\
100 & -110
\end{array}
\right] \qquad ; \qquad
B =
\left[
\begin{array}{c}
10 \\
5
\end{array}
\right]$$

$$ A I = B $$

Entonces:
$$ I = A^{-1} B $$

```python
A = np.array([ [120 , -100] , [100 , -110] ])
B = np.array([ [10] , [5] ])

I = np.dot( np.linalg.inv(A) , B)
```

```python
I
```

```python
#@title Imprime:
print('Resulatados:')
print('I1 = %.4f A'%I[0,0] )
print('I3 = %.4f A'%I[1,0])
print('I2 = %.4f A'%(I[0,0] - I[1,0]) )
```
