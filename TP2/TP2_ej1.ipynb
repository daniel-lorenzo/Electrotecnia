## Problema 1
Dados los siguientes complejos:

$\begin{array}{l}
Z_1 = 6 \\
Z_2 = 2 - j3 \\
Z_3 = j4 \\
Z_4 = -3 + j2 \\
Z_5 = -4 - j4 \\
Z_6 = 3 + j3
\end{array}$ 

### Representarlos en sistema de ejes coordenadas

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import cmath
```

```python
#%matplotlib inline
plt.figure(figsize=(7,7))
re =   [6, 2 , 0, -3,-4,3] 
im =  [0, -3, 4, 2 ,-4,3]
plt.plot(re, im, 'ro')
plt.axis([-5, 8, -5, 5])
plt.title('Números complejos', fontsize=20)
plt.xlabel('Re (Eje real)',  fontsize=14)
plt.ylabel('Im (Eje imaginario)',  fontsize=14)
plt.grid(linestyle="--")
plt.axis([-5,7,-5,5])
plt.show()
```

### Expresarlos en forma Polar y Exponencial

$Z_1 = 6$

#### Solución:

$\left\{
\begin{array}{l}
 |r_1| = \sqrt{(6)^2 + (0)^2} = 6 \\
  \phi_1 = \arctan \frac{Im}{Re} = \arctan \frac{0}{6} = 0^\circ
\end{array}
\right.  
$

Polar:
$$ z_1 = (6 \angle 0^\circ )$$
Exponencial: 
$$ z_1 = 6e^{j0^\circ}$$

En python:

```python
z1 = cmath.polar(6)
```

```python
print('z1 = (%.2f < %.2f°)'%(z1[0], np.degrees(z1[1]) ) )
```

## Representación en coordenadas polares:

```python
lista =[6 + 0j,2-3j,0+4j,-3+2j,-4-4j,3+3j]
radio = []
ang = []
pol = []
for i in lista:
    ang.append( cmath.phase(i) )
    radio.append( abs(i) )
    pol.append(cmath.polar(i))
```

```python
# printing the list using loop 
for x in range( len(radio) ): 
    print('%s = (%.2f < %.2f°)'%(lista[x],radio[x], np.degrees(ang[x]) )) 
```

```python
%matplotlib inline
plt.figure(figsize=(6,6))
for ang, radio in zip(ang, radio):
    plt.polar(ang, radio, 'ro')
plt.grid(linestyle="--")
plt.show()
```
