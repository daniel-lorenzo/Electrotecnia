## Expresar en forma binómica $Z_1 \cdot Z_2$

$Z_1 = 5e^{j \pi/3}$     
$Z_2 = 2e^{-j \pi/6}$

### Solución

$Z_1$ y $Z_2$ están expresados en forma exponencial

En forma polar multiplicamos:
$$ z_1 z_2 = r_1 r_2 ∠(\phi_1 + \phi_2) $$

Reemplazamos:
$$z_1 z_2 = (5)(2) ∠(\pi/3 - \pi/6 )$$

Entonces:
$$z_1 z_2 = 10 ∠ (\pi/6) \; \mathrm{rad} $$

Convertimos a rectangulares:

$$\begin{align}
Re &= r \cos \phi \\
Im &= r \sin \phi
\end{align}$$

Reemplazamos:

$$\begin{align}
Re &= 10 \cos (\pi/6) \approx 8,66 \\
Im &= 10 \sin (\pi/6) = 5 
\end{align}$$

Resulta:
$$z_1 z_2 = 8,66 + j5$$

#### Resolución en python (opcional)

```python
# importamos bibliotecas
import cmath # números complejos
import numpy as np # cálculo numérico
```

```python
# Convertimos a coordenadas rectangulares 
# para almacenarlo en el sistema
z_1 = cmath.rect(5, np.pi/3)
z_2 = cmath.rect(2,-np.pi/6)
```

```python
# Imprime resultados:
print('En forma binómica (coordenadas rectangulares)')
print('z_1*z_2 = {:.2f}'.format(z_1*z_2))
```

## Expresar en forma binómica $Z_1 \cdot Z_2$

$\begin{align}
Z_1 &= 3 + j3 \\
Z_2 &= -3 + j2
\end{align}$

### Solución

Convertimos a polar:
$$\begin{align}
r_1 &= \sqrt{(Re)^2 + (Im)^2}\\
\phi_1 &= \arctan (Im/Re)
\end{align}$$

```python
# importamos bibliotecas
import cmath # números complejos
import numpy as np # cálculo numérico
```

```python
r_1 = np.sqrt(3**2 + 3**2)
# El ángulo se encuentra en el 1er cuadrante
ϕ_1 = np.arctan(3/3)

r_2 = np.sqrt( (-3)**2 + 2**2 )
# El ángulo se encuentra en el 4to cuadrante
ϕ_2 = np.pi - np.arctan(2/3)
```

```python
print('r_1 = %.2f'%r_1)
print('ϕ_1 = %.2f°'%np.rad2deg(ϕ_1))
print('r_2 = %.2f'%r_2)
print('ϕ_2 = %.2f°'%np.rad2deg(ϕ_2))
```

En forma polar multiplicamos:
$$ z_1 z_2 = r_1 r_2 ∠(\phi_1 + \phi_2) $$

Reemplazamos:
$$z_1 z_1 = (4,24)(3,61) ∠ (45^\circ + 146,31^∘)$$

Entonces:
$$z_1 z_2 = 15,30 \angle 191,31^\circ$$

```python
r = r_1*r_2
ϕ = ϕ_1 + ϕ_2
```

```python
print('z_1*z_2 = %.2f < %.2f°'%(r,np.rad2deg(ϕ)))
```

Convertimos a rectangulares:

$$\begin{align}
Re &= r \cos \phi \\
Im &= r \sin \phi
\end{align}$$

```python
Re = r*np.cos(ϕ)
Im = r*np.sin(ϕ)
```

```python
print('Re = %.2f'%Re)
print('Im = %.2f'%Im)
```

En forma binómica, resulta:
$$z_1 z_2 = -15 - j3$$

```python
# Verificamos:
z_1 = 3 + 3j
z_2 = -3 + 2j
```

```python
print('z_1*z_2 = {:.2f}'.format(z_1*z_2))
```
