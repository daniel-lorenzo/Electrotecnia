## Problema 3
Un circuito serie con $R = 25 \, \Omega$ y $L = 0,01 \, \mathrm{Hy}$ se utiliza con frecuencias 50, 100, 500 y 1000 Hz. Calcular las impedancias para cada caso y representar.

### Solución
<div class="alert alert-info">La <strong>impedancia Z</strong> 
 de un circuito es la razón entre la tensión fasorial V y la corriente fasorial I, medida en ohms $(\Omega)$.
    </div>

$$ Z = R + j \omega L $$

$$ Z = R + j 2 \pi f L  $$

$$\begin{array}{lcrcr|lcr}
Z_1 &=& 25 + j 2 \pi (50) \, (0,01)  \, \Omega &=& 25 + j3,14 \ ,\Omega & f_1 &=& 50 \,  \mathrm{Hz} \\
Z_2 &=& 25 + j 2 \pi (100) \, (0,01)  \, \Omega &=& 25 + j6,28 \, \Omega & f_2 &=& 100 \, \mathrm{Hz} \\
Z_3 &=& 25 + j 2 \pi (500) \, (0,01)  \, \Omega &=& 25 + j31,42 \, \Omega & f_3 &=& 500  \, \mathrm{Hz} \\
Z_4 &=& 25 + j 2 \pi (1000) \, (0,01)  \, \Omega &=& 25 + j62,83 \, \Omega & f_4 &=& 1000  \, \mathrm{Hz} \\
\end{array}$$

```python
import math, cmath
```

```python
L = 0.01 # H
R = 25 # Ohm
f1 = 50 # Hz
f2 = 100 # Hz
f3 = 500 # Hz
f4 = 1000 # Hz

Z1 = complex(R , 2*math.pi*f1*L )
Z2 = complex(R , 2*math.pi*f2*L )
Z3 = complex(R , 2*math.pi*f3*L )
Z4 = complex(R , 2*math.pi*f4*L )
```

```python
#@title Resultados:
print('Z1 = {:.2f} Ohm'.format(Z1))
print('Z2 = {:.2f} Ohm'.format(Z2))
print('Z3 = {:.2f} Ohm'.format(Z3))
print('Z4 = {:.2f} Ohm'.format(Z4))
```

```python
import matplotlib.pyplot as plt
```

```python
#@title Gráfico
%matplotlib inline
plt.figure(figsize=(7,7))
ax = plt.axes()
ax = plt.gca()
ax.quiver(0, 0, Z1.real, Z1.imag,angles='xy',scale_units='xy',scale=1,color='red')
ax.quiver(0, 0, Z2.real, Z2.imag,angles='xy',scale_units='xy',scale=1,color='blue')
ax.quiver(0, 0, Z3.real, Z3.imag,angles='xy',scale_units='xy',scale=1,color='green')
ax.quiver(0, 0, Z4.real, Z4.imag,angles='xy',scale_units='xy',scale=1,color='black')
ax.axis([0, 30, 0, 70])
plt.text(Z1.real , Z1.imag, r'$Z_1$', fontsize=18)
plt.text(Z2.real , Z2.imag, r'$Z_2$', fontsize=18)
plt.text(Z3.real , Z3.imag, r'$Z_3$', fontsize=18)
plt.text(Z4.real , Z4.imag, r'$Z_4$', fontsize=18)
plt.xlabel('Re (Eje real)',  fontsize=18)
plt.ylabel('Im (Eje imaginario)',  fontsize=18)
plt.grid(linestyle="--")
ax.set_axisbelow(True)
plt.show()
```
