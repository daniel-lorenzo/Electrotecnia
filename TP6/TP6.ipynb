## Problema 1
Un transformador monofásico ideal tiene un primario de 150 vueltas y un secundario de 750. El primario está conectado a una fuente de 240V y 50Hz. El arrollamiento secundario se conecta a una carga que demanda una corriente de 4A y con un factor de potencia en atraso de 0,8. Determinar:
1. La relación de transformación
2. La corriente en el primario
3. La potencia suministrada a la carga
4. El flujo en el núcleo
5. La tensión por espira

### Solución

### 1) Relación de transformación

__Datos:__
$\left\{
\begin{array}{l}
N_1 = 150 \\
N_2 = 750 \\
\end{array}
\right.$

Siendo:
$$ \frac{V_1}{V_2} = \frac{E_1}{E_2} = \frac{N_1}{N_2} = a $$

```python
# Datos:
N1 = 150
N2 = 750
```

```python
# Cálculo relación de transformación
a = N1/N2
```

```python
#@title Resultado
print('Relación de transformación:')
print('a = %.2f'%a)
```

### 2) Corriente en el primario

__Datos:__ $\rightarrow \; I_2 = 4 \, \mathrm{A}$ 

$$ N_1 I_1 = N_2 I_2 $$

Entonces
$$ I_1 = \frac{N_2}{N_1} I_2 $$

```python
# Dato:
I2 = 4 # A 
```

```python
# Cálculo corriente en el primario
I1 = (N2/N1)*I2
```

```python
#@title Resultado
print('Corriente en el primario')
print('I1 = %.1f A'%I1)
```

### 3) Potencia suministrada a la carga

__Dato:__ $\rightarrow \; \mathrm{fp} = \cos \varphi = 0,8$

$$ P_1 = V_1 I_1 \cos \varphi = P_2 $$

```python
# Datos:
V1 = 240 # V
fp = 0.8
```

```python
# Cálculo potencia suministrada a la carga
P1 = V1*I1*fp
P2 = P1
```

```python
#@title Resultado
print('Potencia suministrada a a la carga:')
print('P2 = %.0f W'%P2)
```

### 4) Flujo en el núcleo
$$ e_1 = \frac{2 \pi}{\sqrt{2}} N_1 f \phi_\mathrm{m \acute ax} $$

$$ \phi_\mathrm{m \acute ax} = e_1 \frac{\sqrt{2}}{2 \pi N_1 f}  $$

```python
# Datos:
e1 = V1
f = 50 # Hz
```

```python
import math
```

```python
# Cálculo flujo en el núcleo
phi_max = V1 * math.sqrt(2)/(2*math.pi*N1*f)
```

```python
#@title Resultado
print('Flujo en el núcleo:')
print('phi_max = %.4f'%phi_max)
```

### 5) Tensión por espira
$$ V_\mathrm{espira} = \frac{V_1}{N_1} = \frac{V_2}{N_2} $$

```python
# Cálculo tensión por espira
V_espira = V1/N1
```

```python
#@title Resultado
print('Tensión por espira:')
print('V_espira = %.2f V'%V_espira)
```

---

```python
#@title Resumen:
print('Relación de transformación')
print('a = %.2f'%a)
print('Corriente en el primario')
print('I1 = %.1f [A]'%I1)
print('Potencia suministrada a la carga')
print('P2 = %.0f [W]'%P2)
print('Flujo en el núcleo')
print('phi_max = %.4f'%phi_max)
print('Tensión por espira')
print('V_espira = %.2f [V]'%V_espira)
```
